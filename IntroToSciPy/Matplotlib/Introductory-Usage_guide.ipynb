# Matplotlib

Las herramientas necesarias para esta sesión la encontraás en este [enlace](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.plot.html#matplotlib.axes.Axes.plot) perteneciente a la documentación de Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Matplotlib grafica nuestros datos en una estructura denominada figura. Cada figura puede contener uno o mas **Axes** (cada región donde hacemos un plot). La forma más sencilla de crear una figura con un Axes es usando `plt.subplots()`, y para agregar datos en dicho Axes usamos `axes_name.plot()`.

En genral tendremos:

* `axes_name.plot([x], y, [fmt], *, data=None, **kwargs)`
* `axes_name.plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)`

```python
# A continuación crearemos una figura en la variable 'fig' 
# con un Axes, donde este último se define en la variable 'ax'
fig,ax = plt.subplots()
```

```python
# A continuación tenemos una figura con 2 Axes
fig,ax = plt.subplots(2)
```

```python
# A continuación tenemos una figura con 2x3 Axes
fig,ax = plt.subplots(2,3)
```

```python
# También podemos crear una figura vacía sin Axes
fig=plt.figure()
```

#### Ejemplo

```python
fig,ax = plt.subplots() 
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

#### Otra forma del ejemplo anterior

Una forma de hacerlo en un solo paso es usando `plt.plot()`, y en el argumento colocar los elementos que representen el domnio y rango.

```python
plt.plot([1,2,3,4],[1,4,2,3])
```

### Más información 

Asimismo, en Matplotlib se denomina **Axis** a los ejes numéricos usados en los Axes, en 2D tendremos 2 Axis, mientras que en 3D, 3. Los límites en los Axis se controlan mediante las siguientes intrucciones:

* `Axes_Name.set_xlim()`
* `Axes_Name.set_ylim()`

Además, cada Axes puede llevar un título, y los Axis etiqueteas, los cuales se fijarán mediante las siguientes instrucciones:

* `Axes_Name.set_title()`
* `Axes_Name.set_xlabel()`
* `Axes_Name.set_ylabel()`

#### Ejemplo

```python
# Definimos el dominio usando NumPy, donde creamos una lista
# con 100 elementos que van de 0 a 2 
x=np.linspace(0,2,100)
x
```

```python
len(x)
```

```python
# A continuación creamos una Figura con un Axes
fig,ax=plt.subplots()
```

```python
# Realizar el gráfico de la función lineal 'y=x'
fig,ax=plt.subplots()
ax.plot(x,x)

# Recordar que debemos poner ambas líneas de código juntas
# sino no tendremos efecto alguno
```

```python
# Ahora agregaremos el título a nuestro Axe, así omo las 
# etiquetas de nuestros Axis
fig,ax=plt.subplots()
ax.plot(x,x)
ax.set_title('Gráfico 1')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
```

Como ya sabemos, podemos tener más de una curva en el mismo Axes, de igual forma podemos agregar legendas dentro de las instrucciones `Axes_Nombre.plot()`, como 'keyword arguments', las cuales serán impresas mediante la instrucción `Axes_Name.legend()`.

```python
fig,ax=plt.subplots()
ax.plot(x,x,label='lineal')
ax.plot(x,x**3,label='cúbica')
ax.set_title('Gráfico 1')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.legend()
```

Podemos hacerlo de otra forma, colocando ambas curvas en una misma línea, sin embargo, tendremos que colocar las etiquetas en la isntrucción `Axes_Name.legend([])`.

```python
fig,ax=plt.subplots()
ax.plot(x,x,x,x**3)
ax.set_title('Gráfico 1')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.legend(['lineal','cúbica'])
```

Podemos manipular a nuestro propio gusto el rango de valores a mostrar en los ejes 'x' e 'y' de nuestro Axes.

```python
fig,ax=plt.subplots()
ax.plot(x,x,x,x**3)
ax.set_title('Gráfico 1')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.legend(['lineal','cúbica'])
ax.set_xlim(-4,4)
ax.set_ylim()
```

También es posible modificar algunas características tales como el color de las curvas, insertar marcadores, el ancho de las líneas o las marcadores. Esto se puede lograr de dos formas:

1. Parámetro [fmt], el cual es un string que resume lo que deseamos agregar mediante letras y símbolos específicos, y luego los keyword arguments.
2. Usando sólo keyword arguments.

#### Forma 1

* Ejemplo 1

```python
# Por motivos visuales modificaré el dominio
x=np.linspace(0,2,10)

fig,ax=plt.subplots()
ax.plot(x,x,'b^',x,x**3,'r--',markersize=8,linewidth=5)
# Como podemos notar 'marksize' se correponde con el marcador 
# usado para la recta 'y=x'.
# De forma anbáloga, 'linewidth' se corresponde con la línea
# usada para la curva 'y=x***3'.

ax.set_title('Gráfico 1')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')

ax.legend(['lineal','cúbica'])

ax.set_xlim(-4,4)
```

* Ejemplo 2

```python
x=np.linspace(0,4,10)

fig,ax=plt.subplots()
ax.plot(x,x,'bo-.',x,x**3,'rd--',ms=10,lw=2)
# Como podemos notar 'marksize'='ms' y 'linewidth'='lw' afectan 
# a ambos; ya que, ahora ambos usan dichas características.

ax.set_title('Gráfico 1')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')

ax.legend(['lineal','cúbica'])

ax.set_xlim(-4,6)
```

* Ejemplo 3

```python
x=np.linspace(0,4,10)

# Independientes
fig,ax=plt.subplots()
ax.plot(x,x,'bo-.','rd--',ms=5,lw=2)
ax.plot(x,x**3,'rd--',ms=10,lw=5)

ax.set_title('Gráfico 1')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')

ax.legend(['lineal','cúbica'])

ax.set_xlim(-4,6)
```

### Forma 2

```python
x=np.linspace(0,4,10)

fig,ax=plt.subplots()
ax.plot(x,x**3,color='orange',ls='-.',ms=8,lw=5,
        label='cúbica')
ax.plot(x,x,color='blue',ls='dotted',marker='^',mec='green',
        mfc='yellow',ms='12',label='lineal')

ax.set_title('Gráfico 1')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')

ax.legend()
        
ax.set_xlim(-1,5)
```

### Varios Axes en una Figura

```python
t=np.linspace(0,5,100)
ran = np.random.randn(len(t))

f1 = np.sin(2*t*np.pi) + ran
f2 = np.cos(2*t*np.pi)-1*t

fig, axs = plt.subplots(2, 1)
axs[0].plot(t, f1, t, f2)
axs[0].set_xlim(-1, 6)
axs[0].set_ylim(-6, 5)
axs[0].set_xlabel('Eje x-1')
axs[0].set_ylabel('Eje y-1')
axs[0].grid(True)

f = axs[1].plot(t, f1+f2, t, f2-f1)
axs[1].set_xlabel('Eje x-2')
axs[1].set_ylabel('Eje y-2')
axs[1].set_xlim(-1, 6)
axs[1].set_ylim(-8, 5)
axs[1].grid(True)

fig.tight_layout()
plt.show()
```
