### Pyplot

Además aprenderemos como insertar código LaTeX y usar su fuente para nuestras etiquetas, títulos, etc.

```python
# Bastante similar a lo anterior, pero ahora usando Pyplot

import matplotlib
import matplotlib.pyplot as plt
from numpy import *
import numpy as np

# Si tenemos LaTeX instalado en nuestras computadoras basta 
# con usar estas líneas de código
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['font.family'] = 'STIXGeneral'

#*********** Si no tenemos instalado LaTeX *****************
# Cambiamos la fuene por defecto a la fuente usada en LaTeX
# esto es DE-TER-MI-NAN-TE para que nuestros plots.
#matplotlib.rcParams['mathtext.fontset'] = 'cm'
#matplotlib.rcParams['font.family'] = 'STIXGeneral'

x=np.linspace(0,4,10)

# Cambiamos el tamaño de la Figura
plt.figure(figsize=(10,4))


#***** Importante ******
# Siempre que usemos código Latex y tengamos el símbolo '\' debemos
# usar una 'r' antes de las comillas del string, tal como veremos 
# a continuación en los ejemplos de este notebook
plt.subplot() 
plt.plot([1,2,3,4],[3,7,8,1],color='blue',ls='-.',marker='^',
        ms=10,lw=3,label=r'Ejemplo $\alpha$')
plt.plot(x,-np.cos(x)*x**2,color='red',ls='dotted',lw=3,
         label='Ejemplo 2')

# Si queremos definir los límites de 'x' e 'y' podemos hacer de
# esta forma. No funciona solo para eje, sino para los dos.
plt.axis([-2,6,-2,16])

plt.ylabel('Some numbers')
plt.xlabel('Este es mi dominio $x$',fontsize=20,color='red')
plt.title('Title')
plt.legend()
plt.grid()
plt.tight_layout() #-> Para que todo quede bien encajado, y nada quede cortado o fuera de la figura
# No usar plt.show() -> Genera error: <Figure size 432x288 with 0 Axes>
plt.savefig('fig1.pdf')
```

### Multiples Figuras y Axes

```python
import matplotlib
import matplotlib.pyplot as plt
from numpy import *
import numpy as np

# Si tenemos LaTeX instalado en nuestras computadoras basta 
# con usar estas líneas de código
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['font.family'] = 'STIXGeneral'

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(figsize=(20,10))

#****** Axes 1 ******
# El siguiente Axes tiene el núermo (221), lo cual significa
# que toma la estructura 2x2 y se ubica en el 1er cuadro
# *********
# * 1 * 2 *
# *********
# * 3 * 4 *
# *********
plt.subplot(221) 
plt.plot(t1, np.exp(-t1) * np.cos(2*np.pi*t1), 'bo', 
         t2, np.exp(-t2) * np.cos(2*np.pi*t2), 'k')
plt.xlabel(r'Etiqueta del eje $x$, una cosita: $\frac{x^9}{4}$',fontsize=20,x=0)
plt.grid()

#****** Axes 2 ******
# De forma análoga este Axes, con núermo (221), toma la estructura
# 2x2 y se ubica en el 3er cuadro
plt.subplot(223)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.grid()

#****** Axes 3 ******
# De forma análoga este Axes, con núermo (122), toma la estructura
# (1x2), es decir, una fila y dos columnas, de tal forma que se
# ubicará en la 2da columna (2do cuadro)
plt.subplot(122) 
plt.plot(t1, np.exp(-t1) * np.cos(2*np.pi*t1), 'bo', 
         t2, np.exp(-t2) * np.cos(2*np.pi*t2), 'k')
plt.grid()
# Agregamos una ecuación sin sentido, sólo para que vean 
# que se puede: plt.text(coord_x,coord_y,keyword-arguments)
plt.text(.5,0.5,r'$\omega=\displaystyle\sum_{x=1}^{n=\infty}\frac{4}{5}\left[e^{-5x}\right]+\sin(x)$',
         fontsize=40,color='b')
plt.title('ABC vs $\mathcal{ABC}$',x=0.5,y=1.1,fontsize=30)
plt.tight_layout()
plt.savefig('fig2.pdf')
```

### Usamos 'Annotating text'

La documentación la encuentras en este [enlace](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.annotate.html#matplotlib.pyplot.annotate)

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure(figsize=(8,4))

plt.plot(t1, np.exp(-t1) * np.cos(2*np.pi*t1), 'bo', 
         t2, np.exp(-t2) * np.cos(2*np.pi*t2), 'k')
plt.xlabel(r'Etiqueta del eje $x$, una cosita: $\frac{x^9}{7}$',
           fontsize=15,x=0)

# Aquí tenemos la forma como debemos usar 'plt.annotate'
plt.annotate('Apunte 1', xy=(1.12,0.27),xytext=(2,.5),
             arrowprops=dict(arrowstyle='<|-|>',
             color='green'),fontsize=20,color='b')
plt.annotate('Apunte 2', xy=(4,0),xytext=(4,0.5),
             arrowprops=dict(arrowstyle='-[',fc='r',
             facecolor='green'),fontsize=14,color='r')
plt.annotate('Apunte 3', xy=(3,0.09),xytext=(3,0.5),
             arrowprops=dict(fc='w',ec='b',width=10,
             headwidth=40,headlength=26,shrink=0.1)
             ,fontsize=24,color='orange')

plt.grid()
plt.tight_layout()
plt.savefig('fig3.pdf')
```
